```python
#null 확인
data.isna().sum()
```

```python
target = data['Survived']
#Sex 문자를 숫자로 변환
data['Sex'] = data['Sex'].map({'male':0, 'female':1})
```

```python
data
```

```python
#비는 값들 평균값으로 대체
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].value_counts().index[0])
```

```python
data_scaled = data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
data_scaled
```

```python
ps_input  = data_scaled[["Pclass", "Sex", "Age", "Fare"]].to_numpy()
ps_target = data_scaled[["Survived"]].to_numpy()
```

```python
train_input, test_input, train_target, test_target = train_test_split(ps_input, ps_target)
```

```python
ss = StandardScaler()
ss.fit(train_input)
```

```python
train_scaled = ss.transform(train_input)
test_scaled  = ss.transform(test_input)
```

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(train_scaled, train_target)
```

```python
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```
